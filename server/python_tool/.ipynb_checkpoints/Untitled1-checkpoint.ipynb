```python
import datetime
# def data_field_exist(col,datafield)
map_type = {
'empty-32-32': "empty",
'random-32-32-20': "random",
'lak303d': "game",
'brc202d': "game",
'empty-8-8': "empty",
'random-64-64-10': "random",
'ost003d': "game",
'den520d': "game",
'maze-128-128-10': "maze",
'ht_mansion_n': "game",
'room-64-64-8': "room",
'empty-48-48': "empty",
'w_woundedcoast': "game",
'Berlin_1_256': "city",
'warehouse-20-40-10-2-1': "warehouse",
'Paris_1_256': "city",
'maze-128-128-1': "maze",
'ht_chantry': "game",
'warehouse-10-20-10-2-1': "warehouse",
'warehouse-20-40-10-2-2': "warehouse",
'maze-128-128-2': "maze",
'random-64-64-20': "random",
'den312d': "game",
'warehouse-10-20-10-2-2': "warehouse",
'Boston_0_256': "city",
'maze-32-32-4': "maze",
'room-32-32-4': "room",
'room-64-64-16': "room",
'lt_gallowstemplar_n': "game",
'orz900d': "game",
'random-32-32-10': "random",
'maze-32-32-2': "maze",
'empty-16-16': "empty",
}
def create_map_info(db):

    mypath = "../dataset/map"
    dir_files = [f for f in listdir(mypath) if isfile(join(mypath, f))]
    col = db["maps"]
    for map_file in dir_files:
        with open("../dataset/map/"+map_file) as myfile:
            head = [next(myfile) for x in range(3)]
        height = head[1].split(" ", 1)[1].split("\n", 1)[0]
        width = head[2].split(" ", 1)[1].split("\n", 1)[0]
        
        split_string = map_file.split(".", 1)
        total_problem = 0; 
        for i in range(1, 26):
            even_scen = "../dataset/scen-even/"+split_string[0]+"-even-"+str(i)+".scen"
            random_scen = "../dataset/scen-random/"+split_string[0]+"-random-"+str(i)+".scen"
            even_num_lines = sum(1 for line in open(even_scen)) - 1
            random_num_lines = sum(1 for line in open(random_scen)) - 1
            total_problem += even_num_lines 
            total_problem += random_num_lines 
        data = {
            "map_name": split_string[0],
            "map_size": width+"x"+height,
            "map_type": map_type[split_string[0]],
            "scens": 50,
            "instances": total_problem,
            "problem_solved": 0,
            "solution_uploaded": 0
        }
        col.insert_one(data)
    col.create_index([("map_name", pymongo.ASCENDING)],unique = True)
        
    
def creat_map_scens(db):
#     url = "mongodb+srv://MAPF:MONASH@mapfbenchmark.zsijqdc.mongodb.net/?retryWrites=true&w=majority"
#     client = pymongo.MongoClient(url)
#     db = client['MAPF_DB']
#     print("Connect successfully")
#     print(db.list_collection_names())


    mypath = "../dataset/map"
    dir_files = [f for f in listdir(mypath) if isfile(join(mypath, f))]
    col = db["scenarios"]
    for map_file in dir_files:
        split_string = map_file.split(".", 1)
        map_id = db["maps"].find_one({"map_name": split_string[0]})['_id']
        for i in range(1, 26):
            even_scen = "../dataset/scen-even/"+split_string[0]+"-even-"+str(i)+".scen"
            random_scen = "../dataset/scen-random/"+split_string[0]+"-random-"+str(i)+".scen"
            even_num_lines = sum(1 for line in open(even_scen)) - 1
            random_num_lines = sum(1 for line in open(random_scen)) - 1
            data = {
                    "map_id": map_id,
                    "scen_type":"even",
                    "type_id":i,
                    "instances":even_num_lines,
                    "problem_solved": 0,
                    "solution_uploaded": 0
            }
            col.insert_one(data)

            data2 = {
                    "map_id": map_id,
                    "scen_type":"random",
                    "type_id":i,
                    "instances":random_num_lines,
                    "problem_solved": 0,
                    "solution_uploaded": 0
            }
            col.insert_one(data2)
    col.create_index([("map_id", pymongo.ASCENDING),("type_id", pymongo.ASCENDING),("scen_type", pymongo.ASCENDING)],unique = True)

    
    
def insert_data_to_leaderboard(db,map_name):
    scen_type = ["even","random"]
    map_id = db["maps"].find_one({"map_name": map_name})['_id']
    for i in range(1, 26):
        for scen_t in scen_type:
            scen_id = db['scenarios'].find_one({"map_id": map_id ,"type_id": i, "scen_type" :  scen_t})['_id']
            number_agent = db['scenarios'].find_one({"map_id": map_id ,"type_id": i, "scen_type" :  scen_t})['instances']
            all_agent_data=[]    
            for a in range(1,number_agent+1):
                instance = {
                    "map_id": map_id,
                    "scen_id": scen_id,
                    "agents": a,
                    "lower_cost": None,
                    "lower_algo_name": None,
                    "lower_algo_id": None,
                    "lower_date":  None,
                    "solution_cost":  None,
                    "solution_algo_name": None,
                    "solution_algo_id": None,
                    "solution_date":  None,
                    "solution_path":  None,
                    "closed": False,
                    "empty": True
                 }
                all_agent_data.append(instance)
            db["instances"].insert_many(all_agent_data)
    db["instances"].create_index([("scen_id", pymongo.ASCENDING),("agents", pymongo.ASCENDING)],unique = True)


            
def create_algo_detail(db):
    CBSH2 = {
        "algo_name": "CBSH2-RTC",
        "authors": "Li et al.",
        "papers": "Jiaoyang Li, Daniel Harabor, Peter J. Stuckey, Hang Ma, Graeme Gange and Sven Koenig. Pairwise Symmetry Reasoning for Multi-Agent Path Finding Search. Artifical Intelligence (AIJ), volume 301, pages 103574, 2021.",
        "github": "https://github.com/Jiaoyang-Li",
        "comments": "",
        'user_id':""
    }
    
    CHBP= {
        "algo_name": "CBSH2-RTC-CHBP",
        "authors": "Shen et al.",
        "papers": "to be inserted",
        "github": "https://github.com/bshen95",
        "comments": "",
        'user_id':""
    }

    db['algorithms'].insert_one(CBSH2)
    db['algorithms'].insert_one(CHBP)

def get_move_string(x,y,p_x,p_y):
    if(x > p_x):
        return "r"
    if(x < p_x):
        return "l"
    if(y > p_y):
        return "u"
    if(y < p_y):
        return "d"
    return "w"
    
def get_path(filename):
    with open(filename) as file:
        planString = ""
        for line in file:
            line  = line[:-3]
            split_string = line.split(":", 1)
            plan = split_string[1].lstrip()
            moves = plan.split("->")
#             fuck it is the reverse output 
            p_y,p_x = map(int, moves[0][1:-1].split(","))
            motionstring = ""
            for move in moves[1:]:
                #             fuck it is the reverse output 
                y,x= map(int, move[1:-1].split(","))
                motionstring   =  motionstring  + get_move_string(x,y,p_x,p_y)                   
                p_x = x 
                p_y = y 
            motionstring = motionstring +"\n"
            planString  = planString + motionstring
        planString = planString.rstrip("\n")
    return planString

def insert_submission(db,submission,instance_id,algo_name,path):
    current_instance = db['instances'].find_one({'_id':instance_id})
    path_id = None
    if(submission['solution_cost']!= None):
        if(current_instance['lower_cost'] != None):
            if(current_instance['lower_cost'] > submission['solution_cost']):
                print("Inserting error lower larger than solution")

#         //insert solution
        if(current_instance['solution_cost'] == None):
            submission["leading_solution"]= True
            submission["best_solution"]= True
            db['instances'].update_one({"_id" : instance_id},{"$set": 
                                                              {"solution_cost": submission['solution_cost'],
                                                                "solution_algo_name": algo_name,
                                                                "solution_algo_id": submission['algo_id'],
                                                                "solution_date":submission['date'],
                                                                "solution_path": path,
                                                                "empty": False
                                                              }
                                                             })
        else:
            if(current_instance['solution_cost'] > submission['solution_cost']):
                db['submissions'].update_many({"instance_id" :  instance_id},{"$set": 
                                                      { 
                                                        "leading_solution": False,
                                                          "best_solution" : False
                                                      }
                                                     })
                submission["leading_solution"]= True
                submission["best_solution"]= True
                db['instances'].update_one({"_id" : instance_id},{"$set": 
                                                      {"solution_cost": submission['solution_cost'],
                                                        "solution_algo_name": algo_name,
                                                        "solution_algo_id": submission['algo_id'],
                                                        "solution_date":submission['date'],
                                                        "solution_path": path,
                                                        "empty": False
                                                      }
                                                     })
            elif (current_instance['solution_cost'] == submission['solution_cost']):
                 submission["best_solution"]= True
                    
    if(submission['lower_cost']!= None):
        if(current_instance['lower_cost'] == None):
            submission["leading_lower"] = True
            submission["best_lower"]= True
            db['instances'].update_one({"_id" : instance_id},{"$set": 
                                                      {"lower_cost": submission['lower_cost'],
                                                        "lower_algo_name": algo_name,
                                                        "lower_algo_id": submission['algo_id'],
                                                        "lower_date":submission['date'],
                                                        "empty": False
                                                      }
                                                     })  
        else:
            if(current_instance['lower_cost']  < submission['lower_cost']):
                db['submissions'].update_many({"instance_id" :  instance_id },{"$set": 
                                          { 
                                            "leading_lower": False,
                                            "best_lower": False
                                          }
                                         })
                submission["leading_lower"] = True
                submission["best_lower"]= True
                db['instances'].update_one({"_id" : instance_id},{"$set": 
                          {"lower_cost": submission['lower_cost'],
                            "lower_algo_name": algo_name,
                            "lower_algo_id": submission['algo_id'],
                            "lower_date":submission['date'],
                            "empty": False
                          }
                         })  
            elif (current_instance['lower_cost']  == submission['lower_cost']):
                submission["best_lower"]= True
                
                
    current_instance = db['instances'].find_one({'_id':instance_id})
    if current_instance['lower_cost'] == current_instance['solution_cost']:
        db['instances'].update_one({"_id" : instance_id},{"$set": 
                          {"closed": True}
                         })  
    db['submissions'].insert_one(submission)
        
        

#         //equal to null

def insert_map_results(db,map_name, algo):
    if algo == "N":
        algo_id = db['algorithms'].find_one({"algo_name":"CBSH2-RTC"})['_id']   
        algo_name = "CBSH2-RTC"
    else:
        algo_id = db['algorithms'].find_one({"algo_name":"CBSH2-RTC-CHBP"})['_id']  
        algo_name = "CBSH2-RTC-CHBP"
    scen_type = ["even","random"]
    map_id = db["maps"].find_one({"map_name": map_name})['_id']
    for i in range(1, 26):
        for scen_t in scen_type: 
            result_file = "../results/"+map_name+"/statistic/"+map_name+"-"+scen_t+"-"+str(i)+"-1800-"+algo+".csv"
            print(result_file)
            df1 = load_data(result_file)
            scen_data =  db['scenarios'].find_one({"map_id": map_id ,"type_id": i, "scen_type" :  scen_t})
            scen_id = scen_data['_id']
            number_agent = scen_data['instances']
            for index, row in df1.iterrows(): 
                if(row['#agents'] > number_agent):
                    break;
                instance_id = db['instances'].find_one({"scen_id": scen_id ,"agents": row['#agents']})['_id']
                if row['solution cost'] == -1:
                    submission= {
                        "map_id": map_id,
                        "instance_id" :  instance_id,
                        "algo_id": algo_id,
                        "lower_cost" : row['min f value'],
                        "solution_cost": None,
                        "date": str(datetime.datetime.now().date()),
                        "best_lower": False,
                        "best_solution": False,
                        "leading_lower": False,
                        "leading_solution": False
                    }
                    path = ""
                else:
                    path_results =  "../results/"+map_name+"/path/"+map_name+"-"+scen_t+"-"+str(i)+"-1800-"+algo+"/"+str(row['#agents']) +"-path.txt"
                    submission= {
                        "map_id": map_id,
                        "instance_id" :  instance_id,
                        "algo_id": algo_id,
                        "lower_cost" : row['solution cost'],
                        "solution_cost": row['solution cost'],
                        "date": str(datetime.datetime.now().date()),
                        "best_lower": False,
                        "best_solution": False,
                        "leading_lower": False,
                        "leading_solution": False
                    }
                    path = get_path(path_results)
                insert_submission(db,submission,instance_id,algo_name,path)

def update_map_progress(db, map_name):
    map_id = db["maps"].find_one({"map_name": map_name})['_id']
    scen_type = ["even","random"]
    total_s = 0
    total_l = 0
    for i in range(1, 26):
        for scen_t in scen_type: 
            scen_id = db['scenarios'].find_one({"map_id": map_id ,"type_id": i, "scen_type" :  scen_t})['_id']
        
            num_s = db["instances"].count_documents({"scen_id" : scen_id, "closed" : True})
            num_l = db["instances"].count_documents({"scen_id" : scen_id, "solution_cost" : {"$ne":None}})
            total_s = total_s + num_s
            total_l = total_l + num_l
            db['scenarios'].update_one({"_id" : scen_id}, {"$set":{
                "problem_solved" : num_s,
                "solution_uploaded" : num_l
            }})
    db['maps'].update_one({"_id" : map_id}, {"$set":{
                "problem_solved" : total_s,
                "solution_uploaded" : total_l
            }})
    
def upload_map_results(db, map_name):
    insert_data_to_leaderboard(db,map_name)
    insert_map_results(db,map_name,"N")
    insert_map_results(db,map_name,"CHBP")
    update_map_progress(db, map_name)
    
def remove_map_records(db, map_name):
    map_id = db["maps"].find_one({"map_name": map_name})['_id']
    db['instances'].delete_many({"map_id":map_id})
    db['submissions'].delete_many({"map_id":map_id})
```

```python
# url = "mongodb+srv://MAPF:MONASH@mapfbenchmark.zsijqdc.mongodb.net/?retryWrites=true&w=majority"
url = "mongodb://localhost:27017/"
client = pymongo.MongoClient(url)

db = client['MAPF_test']
print("Connect successfully")
print(db.list_collection_names())
create_map_info(db)
creat_map_scens(db)
create_algo_detail(db)
insert_data_to_leaderboard(db,"empty-48-48")
insert_map_results(db,"empty-48-48","N")
insert_map_results(db,"empty-48-48","CHBP")
update_map_progress(db, "empty-48-48")
client.close()

import datetime
# def data_field_exist(col,datafield)
map_type = {
'empty-32-32': "empty",
'random-32-32-20': "random",
'lak303d': "game",
'brc202d': "game",
'empty-8-8': "empty",
'random-64-64-10': "random",
'ost003d': "game",
'den520d': "game",
'maze-128-128-10': "maze",
'ht_mansion_n': "game",
'room-64-64-8': "room",
'empty-48-48': "empty",
'w_woundedcoast': "game",
'Berlin_1_256': "city",
'warehouse-20-40-10-2-1': "warehouse",
'Paris_1_256': "city",
'maze-128-128-1': "maze",
'ht_chantry': "game",
'warehouse-10-20-10-2-1': "warehouse",
'warehouse-20-40-10-2-2': "warehouse",
'maze-128-128-2': "maze",
'random-64-64-20': "random",
'den312d': "game",
'warehouse-10-20-10-2-2': "warehouse",
'Boston_0_256': "city",
'maze-32-32-4': "maze",
'room-32-32-4': "room",
'room-64-64-16': "room",
'lt_gallowstemplar_n': "game",
'orz900d': "game",
'random-32-32-10': "random",
'maze-32-32-2': "maze",
'empty-16-16': "empty",
}
def create_map_info(db):

    mypath = "../dataset/map"
    dir_files = [f for f in listdir(mypath) if isfile(join(mypath, f))]
    col = db["maps"]
    print(dir_files)
    for map_file in dir_files:
        with open("../dataset/map/"+map_file) as myfile:
            head = [next(myfile) for x in range(3)]
        height = head[1].split(" ", 1)[1].split("\n", 1)[0]
        width = head[1].split(" ", 1)[1].split("\n", 1)[0]
        
        split_string = map_file.split(".", 1)
        total_problem = 0; 
        for i in range(1, 26):
            even_scen = "../dataset/scen-even/"+split_string[0]+"-even-"+str(i)+".scen"
            random_scen = "../dataset/scen-random/"+split_string[0]+"-random-"+str(i)+".scen"
            even_num_lines = sum(1 for line in open(even_scen)) - 1
            random_num_lines = sum(1 for line in open(random_scen)) - 1
            total_problem += even_num_lines 
            total_problem += random_num_lines 
        data = {
            "map_name": split_string[0],
            "map_size": height+"x"+width,
            "map_type": map_type[split_string[0]],
            "scens": 50,
            "instances": total_problem,
            "problem_solved": 0,
            "solution_uploaded": 0
        }
        col.insert_one(data)
    col.create_index([("map_name", pymongo.ASCENDING)],unique = True)
        
    
def creat_map_scens(db):
#     url = "mongodb+srv://MAPF:MONASH@mapfbenchmark.zsijqdc.mongodb.net/?retryWrites=true&w=majority"
#     client = pymongo.MongoClient(url)
#     db = client['MAPF_DB']
#     print("Connect successfully")
#     print(db.list_collection_names())


    mypath = "../dataset/map"
    dir_files = [f for f in listdir(mypath) if isfile(join(mypath, f))]
    col = db["scenarios"]
    print(dir_files)
    for map_file in dir_files:
        split_string = map_file.split(".", 1)
        map_id = db["maps"].find_one({"map_name": split_string[0]})['_id']
        print(split_string[0])
        for i in range(1, 26):
            even_scen = "../dataset/scen-even/"+split_string[0]+"-even-"+str(i)+".scen"
            random_scen = "../dataset/scen-random/"+split_string[0]+"-random-"+str(i)+".scen"
            even_num_lines = sum(1 for line in open(even_scen)) - 1
            random_num_lines = sum(1 for line in open(random_scen)) - 1
            data = {
                    "map_id": map_id,
                    "scen_id":i,
                    "scen_type":"even",
                    "instances":even_num_lines,
                    "problem_solved": 0,
                    "solution_uploaded": 0
            }
            col.insert_one(data)

            data2 = {
                    "map_id": map_id,
                    "scen_id":i,
                    "scen_type":"random",
                    "instances":random_num_lines,
                    "problem_solved": 0,
                    "solution_uploaded": 0
            }
            col.insert_one(data2)
    col.create_index([("map_id", pymongo.ASCENDING),("scen_id", pymongo.ASCENDING),("scen_type", pymongo.ASCENDING)],unique = True)

    
    
def insert_data_to_leaderboard(db,map_name):
    scen_type = ["even","random"]
    map_id = db["maps"].find_one({"map_name": map_name})['_id']
    for i in range(1, 26):
        for scen_t in scen_type:
            instance_id = db['scenarios'].find_one({"map_id": map_id ,"scen_id": i, "scen_type" :  scen_t})['_id']
            number_agent = db['scenarios'].find_one({"map_id": map_id ,"scen_id": i, "scen_type" :  scen_t})['instances']
            all_agent_data=[]    
            for a in range(1,number_agent+1):
                leader_board = {
                    "instance_id": instance_id,
                    "agents": a,
                    "lower_cost": None,
                    "lower_algo": None,
                    "lower_date":  None,
                    "solution_cost":  None,
                    "solution_algo": None,
                    "solution_date":  None,
                    "solution_path":  None,
                    "closed": False,
                    "empty": True
                 }
                all_agent_data.append(leader_board)
            db["leader_board"].insert_many(all_agent_data)
    

def insert_and_mark_leading(db,data):
    leading = db["submissions"].find_one({
        "instance_id" : data["instance_id"], 
        "agents": data["agents"],
        "type": data["type"], 
        "isleading":True
    })
    if(leading != None):
        #leading exist 
        if db["type"] == "solution":
            if leading['value'] > data['value']:
                db["submission"].update_one({"_id" : leading['_id']},{"$set": {"isleading":False}})
                data['isleading'] = True
            elif leading['value'] == data['value']:
                if leading['runtime'] > data['runtime']:
                    db["submission"].update_one({"_id" : leading['_id']},{"$set": {"isleading":False}})
                    data['isleading'] = True
            db["submissions"].insert_one(data)
        else:
            if leading['value'] < data['value']:
                db["submission"].update_one({"_id" : leading['_id']},{"$set": {"isleading":False}})
                data['isleading'] = True
            elif leading['value'] == data['value']:
                if leading['runtime'] > data['runtime']:
                    db["submission"].update_one({"_id" : leading['_id']},{"$set": {"isleading":False}})
                    data['isleading'] = True
            db["submissions"].insert_one(data)
    else: 
        data['isleading'] = True
        db["submissions"].insert_one(data)
      
        
def get_move_string(x,y,p_x,p_y):
    if(x > p_x):
        return "r"
    if(x < p_x):
        return "l"
    if(y > p_y):
        return "u"
    if(y < p_y):
        return "d"
    return "w"
    
def insert_path_to_data(filename,data):
    with open(filename) as file:
        planString = ""
        for line in file:
            line  = line[:-3]
            split_string = line.split(":", 1)
            plan = split_string[1].lstrip()
            moves = plan.split("->")
            p_x,p_y = map(int, moves[0][1:-1].split(","))
            motionstring = ""
            for move in moves[1:]:
                x,y = map(int, move[1:-1].split(","))
                motionstring   =  motionstring  + get_move_string(x,y,p_x,p_y)                   
                p_x = x 
                p_y = y 
            motionstring = motionstring +"\n"
            planString  = planString + motionstring
        planString = planString.rstrip("\n")
        data['paths'] = planString
    return motionstring
            
def insert_solution(db,data):
    leading = db.find_one({
        "instance_id" : data["instance_id"], 
        "agents": data["agents"],
        "isleading":True
    })
    if(leading != None):
        if leading['value'] > data['value']:
            db.update_one({"_id" : leading['_id']},{"$set": {"isleading":False, "paths":{}}})
            data['isleading'] = True
        else:
            data['paths'] = ""
    else:
        data['isleading'] = True
    db.insert_one(data)

def insert_lower_bound(db,data):
    leading = db.find_one({
        "instance_id" : data["instance_id"], 
        "agents": data["agents"],
        "isleading":True
    })
    if(leading != None):
        if leading['value'] < data['value']:
            db.update_one({"_id" : leading['_id']},{"$set": {"isleading":False}})
            data['isleading'] = True
    else:
        data['isleading'] = True
    db.insert_one(data)
    
def insert_map_results(map_name, algo):
    if algo == "N":
        author_name = "Jiaoyang"
        algo_name = "CBSH2-RTC"
    else:
        author_name = "Bojie"
        algo_name = "CBSH2-RTC + CHBP"
    
    url = "mongodb+srv://MAPF:MONASH@mapfbenchmark.zsijqdc.mongodb.net/?retryWrites=true&w=majority"
    client = pymongo.MongoClient(url)
    db = client['MAPF_DB']
    map_id = db["maps"].find_one({"map_name": map_name})['_id']
    print("Connect successfully")
    scen_type = ["even","random"]
    col = db['instances']
    db["solution_submissions"].create_index([("instance_id", pymongo.ASCENDING),("agents", pymongo.ASCENDING),("type", pymongo.ASCENDING),("isleading", pymongo.ASCENDING)])
    db["lower_bound_submissions"].create_index([("instance_id", pymongo.ASCENDING),("agents", pymongo.ASCENDING),("type", pymongo.ASCENDING),("isleading", pymongo.ASCENDING)])
 
    for i in range(1, 26):
        for scen_t in scen_type: 
            result_file = "../results/"+map_name+"/statistic/"+map_name+"-"+scen_t+"-"+str(i)+"-1800-"+algo+".csv"
            print(result_file)
            df1 = load_data(result_file)
            instance_id = col.find_one({"map_id": map_id ,"scen_id": i, "scen_type" :  scen_t})['_id']

            for index, row in df1.iterrows():
                if row['solution cost'] == -1:
                    lower_bound_submission = {
                        "instance_id" : instance_id,
                        "agents": row['#agents'],
                        "algo_name": algo_name,
                        "authors": author_name,
                        "value": row['min f value'],
                        "runtime": row['runtime'],
                        "isleading": False,
                        "optimal": False,
                        "date": str(datetime.datetime.now().date())                       
                    }
                    insert_lower_bound(db["lower_bound_submissions"],lower_bound_submission)
                else:
                    path_results =  "../results/"+map_name+"/path/"+map_name+"-"+scen_t+"-"+str(i)+"-1800-"+algo+"/"+str(row['#agents']) +"-path.txt"
                    solution_submission = {
                        "instance_id" : instance_id,
                        "agents": row['#agents'],
                        "algo_name": algo_name,
                        "authors": author_name,
                        "value": row['solution cost'],
                        "runtime": row['runtime'],
                        "paths": "",
                        "isleading": False,
                        "optimal": True,
                        "date": str(datetime.datetime.now().date())                       
                    }
                    insert_path_to_data(path_results, solution_submission)
                    insert_solution(db["solution_submissions"],solution_submission)
                    lower_bound_submission = {
                        "instance_id" : instance_id,
                        "agents": row['#agents'],
                        "algo_name": algo_name,
                        "authors": author_name,
                        "value": row['solution cost'],
                        "runtime": row['runtime'],
                        "isleading": False,
                        "optimal": True,
                        "date": str(datetime.datetime.now().date())                       
                    }
                    insert_lower_bound(db["lower_bound_submissions"],lower_bound_submission)
                    
def update_map_progress(map_name):
    url = "mongodb+srv://MAPF:MONASH@mapfbenchmark.zsijqdc.mongodb.net/?retryWrites=true&w=majority"
    client = pymongo.MongoClient(url)
    db = client['MAPF_DB']
    map_id = db["maps"].find_one({"map_name": map_name})['_id']
    print("Connect successfully")
    scen_type = ["even","random"]
    col = db['instances']
    db["solution_submissions"].create_index([("instance_id", pymongo.ASCENDING),("agents", pymongo.ASCENDING),("type", pymongo.ASCENDING),("isleading", pymongo.ASCENDING)])
    db["lower_bound_submissions"].create_index([("instance_id", pymongo.ASCENDING),("agents", pymongo.ASCENDING),("type", pymongo.ASCENDING),("isleading", pymongo.ASCENDING)])
 
    total_s = 0
    total_l = 0
    for i in range(1, 26):
        for scen_t in scen_type: 
            instance_id = col.find_one({"map_id": map_id ,"scen_id": i, "scen_type" :  scen_t})['_id']
            num_s = db["solution_submissions"].count_documents({"instance_id" : instance_id, "isleading" : True})
            num_l = db["lower_bound_submissions"].count_documents({"instance_id" : instance_id, "isleading" : True})
            total_s = total_s + num_s
            total_l = total_l + num_l
            db['instances'].update_one({"_id" : instance_id}, {"$set":{
                "solution_uploaded" : num_s,
                "lower_bound_uploaded" : num_l,
                "upper_bound_uploaded" : num_s,
            }})
    print(total_s)
    print(total_l)
    db['maps'].update_one({"_id" : map_id}, {"$set":{
                "solution_uploaded" : total_s,
                "lower_bound_uploaded" : total_l,
                "upper_bound_uploaded" : total_s,
            }})
    
def upload_map_results(map_name):
    insert_map_results(map_name, "N")
    insert_map_results(map_name, "CHBP")
    update_map_progress(map_name)
```
