```python
def create_algo_detail(db):
    LNS2 = {
        "algo_name": "LNS2",
        "authors": "Li et al.",
        "papers": "Jiaoyang Li, Zhe Chen, Daniel Harabor, Peter J. Stuckey and Sven Koenig. MAPF-LNS2: Fast Repairing for Multi-Agent Path Finding via Large Neighborhood Search In Proceedings of the AAAI Conference on Artificial Intelligence, 2022",
        "github": "https://github.com/Jiaoyang-Li",
        "comments": "",
        'user_id':""
    }

    db['algorithms'].insert_one(LNS2)
    
def get_path(filename):
    with open(filename) as file:
        planString = ""
        for line in file:
            planString  = planString + line
        planString = planString.rstrip("\n")
    return planString

def insert_submission(db,submission,instance_id,algo_name,path):
    current_instance = db['instances'].find_one({'_id':instance_id})
    path_id = None
    if(submission['solution_cost']!= None):
        if(current_instance['lower_cost'] != None):
            if(current_instance['lower_cost'] > submission['solution_cost']):
                print("Inserting error lower larger than solution")

#         //insert solution
        if(current_instance['solution_cost'] == None):
            submission["leading_solution"]= True
            submission["best_solution"]= True
            db['instances'].update_one({"_id" : instance_id},{"$set": 
                                                              {"solution_cost": submission['solution_cost'],
                                                                "solution_algo_name": algo_name,
                                                                "solution_algo_id": submission['algo_id'],
                                                                "solution_date":submission['date'],
                                                                "solution_path": path,
                                                                "empty": False
                                                              }
                                                             })
        else:
            if(current_instance['solution_cost'] > submission['solution_cost']):
                db['submissions'].update_many({"instance_id" :  instance_id},{"$set": 
                                                      { 
                                                        "leading_solution": False,
                                                          "best_solution" : False
                                                      }
                                                     })
                submission["leading_solution"]= True
                submission["best_solution"]= True
                db['instances'].update_one({"_id" : instance_id},{"$set": 
                                                      {"solution_cost": submission['solution_cost'],
                                                        "solution_algo_name": algo_name,
                                                        "solution_algo_id": submission['algo_id'],
                                                        "solution_date":submission['date'],
                                                        "solution_path": path,
                                                        "empty": False
                                                      }
                                                     })
            elif (current_instance['solution_cost'] == submission['solution_cost']):
                 submission["best_solution"]= True                
                
    current_instance = db['instances'].find_one({'_id':instance_id})
    if current_instance['lower_cost'] == current_instance['solution_cost']:
        db['instances'].update_one({"_id" : instance_id},{"$set": 
                          {"closed": True}
                         })  
    db['submissions'].insert_one(submission)
        
        

#         //equal to null

def insert_map_results(db,map_name):
    algo_id = db['algorithms'].find_one({"algo_name":"LNS2"})['_id']  
    algo_name = "LNS2"
    scen_type = ["even","random"]
    map_id = db["maps"].find_one({"map_name": map_name})['_id']
    for i in range(1, 26):
        for scen_t in scen_type: 
            result_file = "../results/"+map_name+"/statistic/"+map_name+"-"+scen_t+"-"+str(i)+"-600-LNS.csv"
            print(result_file)
            df1 = load_data(result_file)
            scen_data =  db['scenarios'].find_one({"map_id": map_id ,"type_id": i, "scen_type" :  scen_t})
            scen_id = scen_data['_id']
            number_agent = scen_data['instances']
            for index, row in df1.iterrows(): 
                if(row['#agents'] > number_agent):
                    break;
                instance_id = db['instances'].find_one({"scen_id": scen_id ,"agents": row['#agents']})['_id']
                if row['solution cost'] != -1:
                    path_results =  "../results/"+map_name+"/path/"+map_name+"-"+scen_t+"-"+str(i)+"-600/"+str(row['#agents']) +"-path.txt"
                    submission= {
                        "map_id": map_id,
                        "instance_id" :  instance_id,
                        "algo_id": algo_id,
                        "lower_cost" : None,
                        "solution_cost": row['solution cost'],
                        "date": str(datetime.datetime.now().date()),
                        "best_lower": False,
                        "best_solution": False,
                        "leading_lower": False,
                        "leading_solution": False
                    }
                    path = get_path(path_results)
                insert_submission(db,submission,instance_id,algo_name,path)
    
def update_map_progress(db, map_name):
    map_id = db["maps"].find_one({"map_name": map_name})['_id']
    scen_type = ["even","random"]
    total_s = 0
    total_l = 0
    for i in range(1, 26):
        for scen_t in scen_type: 
            scen_id = db['scenarios'].find_one({"map_id": map_id ,"type_id": i, "scen_type" :  scen_t})['_id']
        
            num_s = db["instances"].count_documents({"scen_id" : scen_id, "closed" : True})
            num_l = db["instances"].count_documents({"scen_id" : scen_id, "solution_cost" : {"$ne":None}})
            total_s = total_s + num_s
            total_l = total_l + num_l
            db['scenarios'].update_one({"_id" : scen_id}, {"$set":{
                "problem_solved" : num_s,
                "solution_uploaded" : num_l
            }})
    db['maps'].update_one({"_id" : map_id}, {"$set":{
                "problem_solved" : total_s,
                "solution_uploaded" : total_l
            }})
    
def remove_algo_records_from_map(db, map_name,algo_name):
    all_algorithms = db['algorithms'].find({})
    algo_mapper = {}
    for algo in all_algorithms:
        print(algo)
        algo_mapper[algo['_id']] = algo['algo_name']
#         algo_mapper[algo['_id']] = algo['algo_name']
    print(algo_mapper)
    map_id = db["maps"].find_one({"map_name": map_name})['_id']
    print(map_id)
    algo_id = db['algorithms'].find_one({"algo_name":algo_name})['_id']    
    cursors = db['instances'].find({"map_id" : map_id,"lower_algo_id":algo_id})
    for cursor in cursors:
        print(cursor.get("_id"))
        
    cursors = db['instances'].find({"map_id" : map_id,"solution_algo_id":algo_id})
    
#     pipeline = [
#        {
#           "$match": {
#              "map_id" : map_id
#           }
#        }, 
#        {'$lookup': 
#                 {'from' : 'submissions',
#                  'localField' : '_id',
#                  'foreignField' : 'instance_id',
#                  'as' : 'submissions'}
#        }
#     ]
    
    
    pipeline = [
                {"$match":{"map_id": map_id,
            "$or": [ { "solution_algo_id":algo_id } ,
                { "lower_algo_id": algo_id} ]
            }},
               {'$lookup': 
                {'from' : 'submissions',
                 'localField' : '_id',
                 'foreignField' : 'instance_id',
                 'as' : "Submission_records"}
            },
            {
                "$project" : {
                    "_id" : "$_id",
                    "map_id": "$map_id",
                    "scen_id":"$scen_id",
                    "agents":"$agents",
                    "Submission_records": {
                        "$filter": {
                            "input": '$Submission_records',
                            "cond": { "$ne": [ "$$submission.algo_id", algo_id] },
                            "as": 'submission',
                        }
                    }
                }
            }
    ]

#     db['instances'].aggregate(pipeline)
#     print("finished")
    for doc in (db['instances'].aggregate(pipeline)):
        instance_id = doc['_id']
        update_instance = {
                    "map_id": doc["map_id"],
                    "scen_id": doc["scen_id"],
                    "agents": doc["agents"],
                    "lower_cost": None,
                    "lower_algo_name": None,
                    "lower_algo_id": None,
                    "lower_date":  None,
                    "solution_cost":  None,
                    "solution_algo_name": None,
                    "solution_algo_id": None,
                    "solution_date":  None,
                    "solution_path":  None,
                    "closed": False,
                    "empty": True
                 }
        if(not doc["Submission_records"]):
            #replace the entire record
            b = 1
        else:
            for submission in doc['Submission_records']:
                if(submission['lower_cost'] != None):
                    if (update_instance['lower_cost'] !=None):
                        if (submission['lower_cost'] >= update_instance['lower_cost']):
                            update_instance['lower_cost'] = submission['lower_cost']
                            update_instance['lower_algo_name'] = algo_mapper[submission['algo_id']]
                            update_instance['lower_algo_id'] = submission['algo_id']
                            update_instance['lower_date'] = submission['date']
                            update_instance['empty'] = False
                        elif (submission['lower_cost'] == update_instance['lower_cost']):
                            if(submission['date'] < update_instance['lower_date']):
                                print(update_instance['lower_date'])
                                print(submission['date'])
                                update_instance['lower_cost'] = submission['lower_cost']
                                update_instance['lower_algo_name'] = algo_mapper[submission['algo_id']]
                                update_instance['lower_algo_id'] = submission['algo_id']
                                update_instance['lower_date'] = submission['date']
                                update_instance['empty'] = False
                    else: 
                        update_instance['lower_cost'] = submission['lower_cost']
                        update_instance['lower_algo_name'] = algo_mapper[submission['algo_id']]
                        update_instance['lower_algo_id'] = submission['algo_id']
                        update_instance['lower_date'] = submission['date']
                        update_instance['empty'] = False
            for submission in doc['Submission_records']:
                need_update = False 
                leading_lower = False
                best_lower = False
                if (submission['lower_cost'] == update_instance['lower_cost'] and submission['best_lower'] == False):
                    need_update = True
                    best_lower = True
                if (submission['algo_id'] == update_instance['lower_algo_id']  and submission['leading_lower'] == False):
                    need_update = True
                    leading_lower = True
                if(need_update):
                    for ss in doc['Submission_records']:
                        print(ss)
#                     pprint(doc)
#                     pprint(update_instance)
#                     pprint(submission)
                    pprint('impossible ?')
                    pprint('impossible ?')
                    pprint('impossible ?')
                    pprint('impossible ?')
                    pprint('impossible ?')
                    
#         pprint(doc)
#     for cursor in cursors:
#         instance_id = cursor.get("_id")
#         other_submissions = db['submissions'].find({"instance_id" : instance_id, "algo_id": {"$ne": algo_id }, "solution_cost":{"$ne":None}})
#         for sub in other_submissions:
#             print(sub.get("_id"))
#             print(cursor.get("solution_cost"))
#             print(sub.get("solution_cost"))
    
#     db['submissions'].delete_many({"map_id":map_id, "algo_id": algo_id })
```
