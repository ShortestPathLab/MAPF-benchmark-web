```python
import datetime
# def data_field_exist(col,datafield)
map_type = {
'empty-32-32': "empty",
'random-32-32-20': "random",
'lak303d': "game",
'brc202d': "game",
'empty-8-8': "empty",
'random-64-64-10': "random",
'ost003d': "game",
'den520d': "game",
'maze-128-128-10': "maze",
'ht_mansion_n': "game",
'room-64-64-8': "room",
'empty-48-48': "empty",
'w_woundedcoast': "game",
'Berlin_1_256': "city",
'warehouse-20-40-10-2-1': "warehouse",
'Paris_1_256': "city",
'maze-128-128-1': "maze",
'ht_chantry': "game",
'warehouse-10-20-10-2-1': "warehouse",
'warehouse-20-40-10-2-2': "warehouse",
'maze-128-128-2': "maze",
'random-64-64-20': "random",
'den312d': "game",
'warehouse-10-20-10-2-2': "warehouse",
'Boston_0_256': "city",
'maze-32-32-4': "maze",
'room-32-32-4': "room",
'room-64-64-16': "room",
'lt_gallowstemplar_n': "game",
'orz900d': "game",
'random-32-32-10': "random",
'maze-32-32-2': "maze",
'empty-16-16': "empty",
}
def create_map_info():
    url = "mongodb+srv://MAPF:MONASH@mapfbenchmark.zsijqdc.mongodb.net/?retryWrites=true&w=majority"
    client = pymongo.MongoClient(url)
    db = client['MAPF_DB']
    print("Connect successfully")
    print(db.list_collection_names())

    mypath = "../dataset/map"
    dir_files = [f for f in listdir(mypath) if isfile(join(mypath, f))]
    col = db["maps"]
    print(dir_files)
    for map_file in dir_files:
        with open("../dataset/map/"+map_file) as myfile:
            head = [next(myfile) for x in range(3)]
        height = head[1].split(" ", 1)[1].split("\n", 1)[0]
        width = head[1].split(" ", 1)[1].split("\n", 1)[0]
        
        split_string = map_file.split(".", 1)
        total_problem = 0; 
        for i in range(1, 26):
            even_scen = "../dataset/scen-even/"+split_string[0]+"-even-"+str(i)+".scen"
            random_scen = "../dataset/scen-random/"+split_string[0]+"-random-"+str(i)+".scen"
            even_num_lines = sum(1 for line in open(even_scen)) - 1
            random_num_lines = sum(1 for line in open(random_scen)) - 1
            total_problem += even_num_lines 
            total_problem += random_num_lines 
        data = {
            "map_name": split_string[0],
            "map_size": height+"x"+width,
            "map_type": map_type[split_string[0]],
            "instances": 50,
            "problems": total_problem,
            "solution_uploaded": 0,
            "lower_bound_uploaded": 0,
            "upper_bound_uploaded": 0
        }
        col.insert_one(data)
    col.create_index([("map_name", pymongo.ASCENDING)],unique = True)
    client.close()
        
    
def creat_map_instances():
    url = "mongodb+srv://MAPF:MONASH@mapfbenchmark.zsijqdc.mongodb.net/?retryWrites=true&w=majority"
    client = pymongo.MongoClient(url)
    db = client['MAPF_DB']
    print("Connect successfully")
    print(db.list_collection_names())


    mypath = "../dataset/map"
    dir_files = [f for f in listdir(mypath) if isfile(join(mypath, f))]
    col = db["instances"]
    print(dir_files)
    for map_file in dir_files:
        split_string = map_file.split(".", 1)
        map_id = db["maps"].find_one({"map_name": split_string[0]})['_id']
        print(split_string[0])
        for i in range(1, 26):
            even_scen = "../dataset/scen-even/"+split_string[0]+"-even-"+str(i)+".scen"
            random_scen = "../dataset/scen-random/"+split_string[0]+"-random-"+str(i)+".scen"
            even_num_lines = sum(1 for line in open(even_scen)) - 1
            random_num_lines = sum(1 for line in open(random_scen)) - 1
            data = {
                    "map_id": map_id,
                    "scen_id":i,
                    "scen_type":"even",
                    "problems":even_num_lines,
                    "solution_uploaded": 0,
                    "lower_bound_uploaded": 0,
                    "upper_bound_uploaded": 0
            }
            col.insert_one(data)

            data2 = {
                    "map_id": map_id,
                    "scen_id":i,
                    "scen_type":"random",
                    "problems":random_num_lines,
                    "solution_uploaded": 0,
                    "lower_bound_uploaded": 0,
                    "upper_bound_uploaded": 0
            }
            col.insert_one(data2)
    col.create_index([("map_id", pymongo.ASCENDING),("scen_id", pymongo.ASCENDING),("scen_type", pymongo.ASCENDING)],unique = True)
    client.close()

    
    
# def insert_data_to_leaderboard(db,data,submission_type):
#     db_submission = db["submissions"]
#     db_leader = db["leaderboard"]
#     result = db_leader.find_one({"instance_id": instance_id, "agents": row['#agents'], "type": submission_type})
#     if list(result):
#         if submission_type == "lower_bound":
#             if result['cost'] < data['cost']:
            
#             else:

def insert_and_mark_leading(db,data):
    leading = db["submissions"].find_one({
        "instance_id" : data["instance_id"], 
        "agents": data["agents"],
        "type": data["type"], 
        "isleading":True
    })
    if(leading != None):
        #leading exist 
        if db["type"] == "solution":
            if leading['value'] > data['value']:
                db["submission"].update_one({"_id" : leading['_id']},{"$set": {"isleading":False}})
                data['isleading'] = True
            elif leading['value'] == data['value']:
                if leading['runtime'] > data['runtime']:
                    db["submission"].update_one({"_id" : leading['_id']},{"$set": {"isleading":False}})
                    data['isleading'] = True
            db["submissions"].insert_one(data)
        else:
            if leading['value'] < data['value']:
                db["submission"].update_one({"_id" : leading['_id']},{"$set": {"isleading":False}})
                data['isleading'] = True
            elif leading['value'] == data['value']:
                if leading['runtime'] > data['runtime']:
                    db["submission"].update_one({"_id" : leading['_id']},{"$set": {"isleading":False}})
                    data['isleading'] = True
            db["submissions"].insert_one(data)
    else: 
        data['isleading'] = True
        db["submissions"].insert_one(data)

def get_move_string(x,y,p_x,p_y):
    if(x > p_x):
        return "r"
    if(x < p_x):
        return "l"
    if(y > p_y):
        return "u"
    if(y < p_y):
        return "d"
    return "w"
    
def insert_path_to_data(filename,data):
    with open(filename) as file:
        index = 0
        for line in file:
            line  = line[:-3]
            split_string = line.split(":", 1)
            plan = split_string[1].lstrip()
            moves = plan.split("->")
            p_x,p_y = map(int, moves[0][1:-1].split(","))
            motionstring = ""
            a = 0
            for move in moves[1:]:
                x,y = map(int, move[1:-1].split(","))
                motionstring   =  motionstring  + get_move_string(x,y,p_x,p_y)                   
                p_x = x 
                p_y = y 
            print(motionstring)
            data['paths'][str(index)] = split_string[1].lstrip()
            index = index + 1 
            
def insert_solution(db,data):
    leading = db.find_one({
        "instance_id" : data["instance_id"], 
        "agents": data["agents"],
        "isleading":True
    })
    if(leading != None):
        if leading['value'] > data['value']:
            db.update_one({"_id" : leading['_id']},{"$set": {"isleading":False, "paths":{}}})
            data['isleading'] = True
        else:
            data['paths'] = {}
    else:
        data['isleading'] = True
    db.insert_one(data)

def insert_lower_bound(db,data):
    leading = db.find_one({
        "instance_id" : data["instance_id"], 
        "agents": data["agents"],
        "isleading":True
    })
    if(leading != None):
        if leading['value'] < data['value']:
            db.update_one({"_id" : leading['_id']},{"$set": {"isleading":False}})
            data['isleading'] = True
    else:
        data['isleading'] = True
    db.insert_one(data)
    
def insert_map_results(map_name, algo):
    if algo == "N":
        author_name = "Jiaoyang"
        algo_name = "CBSH2-RTC"
    else:
        author_name = "Bojie"
        algo_name = "CBSH2-RTC + CHBP"
    
    url = "mongodb+srv://MAPF:MONASH@mapfbenchmark.zsijqdc.mongodb.net/?retryWrites=true&w=majority"
    client = pymongo.MongoClient(url)
    db = client['MAPF_DB']
    map_id = db["maps"].find_one({"map_name": map_name})['_id']
    print("Connect successfully")
    scen_type = ["even","random"]
    col = db['instances']
    db["solution_submissions"].create_index([("instance_id", pymongo.ASCENDING),("agents", pymongo.ASCENDING),("type", pymongo.ASCENDING),("isleading", pymongo.ASCENDING)])
    db["lower_bound_submissions"].create_index([("instance_id", pymongo.ASCENDING),("agents", pymongo.ASCENDING),("type", pymongo.ASCENDING),("isleading", pymongo.ASCENDING)])
 
    for i in range(1, 26):
        for scen_t in scen_type: 
            result_file = "../results/"+map_name+"/statistic/"+map_name+"-"+scen_t+"-"+str(i)+"-1800-"+algo+".csv"
            print(result_file)
            df1 = load_data(result_file)
            instance_id = col.find_one({"map_id": map_id ,"scen_id": i, "scen_type" :  scen_t})['_id']

            for index, row in df1.iterrows():
                if row['solution cost'] == -1:
                    lower_bound_submission = {
                        "instance_id" : instance_id,
                        "agents": row['#agents'],
                        "algo_name": algo_name,
                        "authors": author_name,
                        "value": row['min f value'],
                        "runtime": row['runtime'],
                        "isleading": False,
                        "optimal": False,
                        "date": str(datetime.datetime.now().date())                       
                    }
#                     insert_lower_bound(db["lower_bound_submissions"],lower_bound_submission)
                else:
                    path_results =  "../results/"+map_name+"/path/"+map_name+"-"+scen_t+"-"+str(i)+"-1800-"+algo+"/"+str(row['#agents']) +"-path.txt"
                    solution_submission = {
                        "instance_id" : instance_id,
                        "agents": row['#agents'],
                        "algo_name": algo_name,
                        "authors": author_name,
                        "value": row['solution cost'],
                        "runtime": row['runtime'],
                        "paths": {},
                        "isleading": False,
                        "optimal": True,
                        "date": str(datetime.datetime.now().date())                       
                    }
                    insert_path_to_data(path_results, solution_submission)
#                     insert_solution(db["solution_submissions"],solution_submission)
                    lower_bound_submission = {
                        "instance_id" : instance_id,
                        "agents": row['#agents'],
                        "algo_name": algo_name,
                        "authors": author_name,
                        "value": row['solution cost'],
                        "runtime": row['runtime'],
                        "isleading": False,
                        "optimal": True,
                        "date": str(datetime.datetime.now().date())                       
                    }
#                     insert_lower_bound(db["lower_bound_submissions"],lower_bound_submission)
                    
def update_map_progress(map_name):
    url = "mongodb+srv://MAPF:MONASH@mapfbenchmark.zsijqdc.mongodb.net/?retryWrites=true&w=majority"
    client = pymongo.MongoClient(url)
    db = client['MAPF_DB']
    map_id = db["maps"].find_one({"map_name": map_name})['_id']
    print("Connect successfully")
    scen_type = ["even","random"]
    col = db['instances']
    db["solution_submissions"].create_index([("instance_id", pymongo.ASCENDING),("agents", pymongo.ASCENDING),("type", pymongo.ASCENDING),("isleading", pymongo.ASCENDING)])
    db["lower_bound_submissions"].create_index([("instance_id", pymongo.ASCENDING),("agents", pymongo.ASCENDING),("type", pymongo.ASCENDING),("isleading", pymongo.ASCENDING)])
 
    total_s = 0
    total_l = 0
    for i in range(1, 26):
        for scen_t in scen_type: 
            instance_id = col.find_one({"map_id": map_id ,"scen_id": i, "scen_type" :  scen_t})['_id']
            num_s = db["solution_submissions"].count_documents({"instance_id" : instance_id, "isleading" : True})
            num_l = db["lower_bound_submissions"].count_documents({"instance_id" : instance_id, "isleading" : True})
            total_s = total_s + num_s
            total_l = total_l + num_l
            db['instances'].update_one({"_id" : instance_id}, {"$set":{
                "solution_uploaded" : num_s,
                "lower_bound_uploaded" : num_l,
                "upper_bound_uploaded" : num_s,
            }})
    print(total_s)
    print(total_l)
    db['maps'].update_one({"_id" : map_id}, {"$set":{
                "solution_uploaded" : total_s,
                "lower_bound_uploaded" : total_l,
                "upper_bound_uploaded" : total_s,
            }})
    
def upload_map_results(map_name):
    insert_map_results(map_name, "N")
    insert_map_results(map_name, "CHBP")
    update_map_progress(map_name)
    
    
```

```python
import datetime
# def data_field_exist(col,datafield)
map_type = {
'empty-32-32': "empty",
'random-32-32-20': "random",
'lak303d': "game",
'brc202d': "game",
'empty-8-8': "empty",
'random-64-64-10': "random",
'ost003d': "game",
'den520d': "game",
'maze-128-128-10': "maze",
'ht_mansion_n': "game",
'room-64-64-8': "room",
'empty-48-48': "empty",
'w_woundedcoast': "game",
'Berlin_1_256': "city",
'warehouse-20-40-10-2-1': "warehouse",
'Paris_1_256': "city",
'maze-128-128-1': "maze",
'ht_chantry': "game",
'warehouse-10-20-10-2-1': "warehouse",
'warehouse-20-40-10-2-2': "warehouse",
'maze-128-128-2': "maze",
'random-64-64-20': "random",
'den312d': "game",
'warehouse-10-20-10-2-2': "warehouse",
'Boston_0_256': "city",
'maze-32-32-4': "maze",
'room-32-32-4': "room",
'room-64-64-16': "room",
'lt_gallowstemplar_n': "game",
'orz900d': "game",
'random-32-32-10': "random",
'maze-32-32-2': "maze",
'empty-16-16': "empty",
}
def create_map_info():
    url = "mongodb+srv://MAPF:MONASH@mapfbenchmark.zsijqdc.mongodb.net/?retryWrites=true&w=majority"
    client = pymongo.MongoClient(url)
    db = client['MAPF_DB']
    print("Connect successfully")
    print(db.list_collection_names())

    mypath = "../dataset/map"
    dir_files = [f for f in listdir(mypath) if isfile(join(mypath, f))]
    col = db["maps"]
    print(dir_files)
    for map_file in dir_files:
        with open("../dataset/map/"+map_file) as myfile:
            head = [next(myfile) for x in range(3)]
        height = head[1].split(" ", 1)[1].split("\n", 1)[0]
        width = head[1].split(" ", 1)[1].split("\n", 1)[0]
        
        split_string = map_file.split(".", 1)
        total_problem = 0; 
        for i in range(1, 26):
            even_scen = "../dataset/scen-even/"+split_string[0]+"-even-"+str(i)+".scen"
            random_scen = "../dataset/scen-random/"+split_string[0]+"-random-"+str(i)+".scen"
            even_num_lines = sum(1 for line in open(even_scen)) - 1
            random_num_lines = sum(1 for line in open(random_scen)) - 1
            total_problem += even_num_lines 
            total_problem += random_num_lines 
        data = {
            "map_name": split_string[0],
            "map_size": height+"x"+width,
            "map_type": map_type[split_string[0]],
            "instances": 50,
            "problems": total_problem,
            "solution_uploaded": 0,
            "lower_bound_uploaded": 0,
            "upper_bound_uploaded": 0
        }
        col.insert_one(data)
    col.create_index([("map_name", pymongo.ASCENDING)],unique = True)
    client.close()
        
    
def creat_map_instances():
    url = "mongodb+srv://MAPF:MONASH@mapfbenchmark.zsijqdc.mongodb.net/?retryWrites=true&w=majority"
    client = pymongo.MongoClient(url)
    db = client['MAPF_DB']
    print("Connect successfully")
    print(db.list_collection_names())


    mypath = "../dataset/map"
    dir_files = [f for f in listdir(mypath) if isfile(join(mypath, f))]
    col = db["instances"]
    print(dir_files)
    for map_file in dir_files:
        split_string = map_file.split(".", 1)
        map_id = db["maps"].find_one({"map_name": split_string[0]})['_id']
        print(split_string[0])
        for i in range(1, 26):
            even_scen = "../dataset/scen-even/"+split_string[0]+"-even-"+str(i)+".scen"
            random_scen = "../dataset/scen-random/"+split_string[0]+"-random-"+str(i)+".scen"
            even_num_lines = sum(1 for line in open(even_scen)) - 1
            random_num_lines = sum(1 for line in open(random_scen)) - 1
            data = {
                    "map_id": map_id,
                    "scen_id":i,
                    "scen_type":"even",
                    "problems":even_num_lines,
                    "solution_uploaded": 0,
                    "lower_bound_uploaded": 0,
                    "upper_bound_uploaded": 0
            }
            col.insert_one(data)

            data2 = {
                    "map_id": map_id,
                    "scen_id":i,
                    "scen_type":"random",
                    "problems":random_num_lines,
                    "solution_uploaded": 0,
                    "lower_bound_uploaded": 0,
                    "upper_bound_uploaded": 0
            }
            col.insert_one(data2)
    col.create_index([("map_id", pymongo.ASCENDING),("scen_id", pymongo.ASCENDING),("scen_type", pymongo.ASCENDING)],unique = True)
    client.close()

    
def add_all_agent_data(map_name):    
    url = "mongodb+srv://MAPF:MONASH@mapfbenchmark.zsijqdc.mongodb.net/?retryWrites=true&w=majority"
    client = pymongo.MongoClient(url)
    db = client['MAPF_DB']
    print("Connect successfully")
    map_id = db["maps"].find_one({"map_name": map_name})['_id']
    col = db['instances']
    scen_type = ["even","random"] 
    all_agent_data=[]    
    for i in range(1, 26):
        for scen_t in scen_type: 
            res = col.find_one({"map_id": map_id,"scen_id": i, "scen_type" :  scen_t})
            number_agent = res['problems']
            instance_id = res['_id']
            for a in range(1,number_agent+1):
                all_agent_data.append(
                    {
                        "instance_id":instance_id,
                        "description": map_name+"-"+ str(i) + "-"+ scen_t,
                        "agents": a,
                        "leading_cost": -1,
                        "leading_runtime": 1800,
                        "leading_algo": "unknow",
                        "leading_author": "unknow",
                        "submissions":[]
                    }
                )
    db['solution_records'].insert_many(all_agent_data)
    db['lower_bound_records'].insert_many(all_agent_data)
    db['upper_bound_records'].insert_many(all_agent_data)    
    print("finish inserting")
    db['solution_records'].create_index([("instance_id", pymongo.ASCENDING),("agents", pymongo.ASCENDING)],unique = True)
    db['lower_bound_records'].create_index([("instance_id", pymongo.ASCENDING),("agents", pymongo.ASCENDING)],unique = True)
    db['upper_bound_records'].create_index([("instance_id", pymongo.ASCENDING),("agents", pymongo.ASCENDING)],unique = True)
    print("finish indexing")
    
def insert_map_results(map_name, algo):
    if algo == "N":
        author_name = "Jiaoyang"
        algo_name = "CBSH2-RTC"
    else:
        author_name = "Bojie"
        algo_name = "CBSH2-RTC + CHBP"
    
    url = "mongodb+srv://MAPF:MONASH@mapfbenchmark.zsijqdc.mongodb.net/?retryWrites=true&w=majority"
    client = pymongo.MongoClient(url)
    db = client['MAPF_DB']
    map_id = db["maps"].find_one({"map_name": map_name})['_id']
    print("Connect successfully")
    scen_type = ["even","random"]
    col = db['instances']
    lower_query = []
    upper_query = []
    solution_query = []    

    for i in range(1, 26):
        for scen_t in scen_type: 
            result_file = "../results/"+map_name+"/statistic/"+map_name+"-"+scen_t+"-"+str(i)+"-1800-"+algo+".csv"
            print(result_file)
            df1 = load_data(result_file)
            instance_id = col.find_one({"map_id": map_id ,"scen_id": i, "scen_type" :  scen_t})['_id']
            for index, row in df1.iterrows():
                if row['solution cost'] == -1:
                    history = {
                        "subproblem_id":-1,
                        "instance_id": instance_id,
                        "agents": row['#agents'],
                        "algo_name": algo_name,
                        "authors": author_name,
                        "cost": row['solution cost'],
                        "runtime": row['runtime'],
                        "date": str(datetime.datetime.now().date())                       
                    }
                    lower_query.append(history)
                else:
                    history = {
                        "subproblem_id":-1,
                        "instance_id": instance_id,
                        "agents": row['#agents'],
                        "algo_name": algo_name,
                        "authors": author_name,
                        "cost": row['solution cost'],
                        "runtime": row['runtime'],
                        "date": str(datetime.datetime.now().date())                       
                    }
                    solution_query.append(history)
                    lower_query.append(history)
                    upper_query.append(history)
                    
    db['lower_bound_submissions'].insert_many(lower_query)
    print("finish inserting lower bound submissions")
    db["solution_submissions"].insert_many(solution_query)
    print("finish inserting solution submissions")
    db["upper_bound_submissions"].insert_many(upper_query)     
    print("finish inserting upper bound submissions")

    
def upload_map_results(map_name):
    add_all_agent_data(map_name)
    insert_map_results(map_name, "N")
    insert_map_results(map_name, "CHBP")
    
    
    
    
    
    
    
    
    
    
    

# def add_all_agent_data(map_name):    
#     url = "mongodb+srv://MAPF:MONASH@mapfbenchmark.zsijqdc.mongodb.net/?retryWrites=true&w=majority"
#     client = pymongo.MongoClient(url)
#     db = client['MAPF_DB']
#     print("Connect successfully")
#     map_id = db["maps"].find_one({"map_name": map_name})['_id']
#     col = db['instances']
#     scen_type = ["even","random"] 
#     all_agent_data=[]    
#     for i in range(1, 26):
#         for scen_t in scen_type: 
#             res = col.find_one({"map_id": map_id,"instance_id": i, "instance_type" :  scen_t})
#             number_agent = res['total_agent']
#             instance_id = res['_id']
#             for a in range(1,number_agent+1):
#                 all_agent_data.append(
#                     {
#                         "map_instance_id":instance_id,
#                         "description": map_name+"-"+ str(i) + "-"+ scen_t,
#                         "#agents": a,
#                         "history":[]
#                     }
#                 )
#     db['soultion_records'].insert_many(all_agent_data)
#     db['lower_bound_records'].insert_many(all_agent_data)
#     db['upper_bound_records'].insert_many(all_agent_data)
    
#     db['lower_bound_records'].create_index([("map_instance_id", pymongo.ASCENDING),("#agents", pymongo.ASCENDING)],unique = True)
#     db['soultion_records'].create_index([("map_instance_id", pymongo.ASCENDING),("#agents", pymongo.ASCENDING)],unique = True)
#     db['upper_bound_records'].create_index([("map_instance_id", pymongo.ASCENDING),("#agents", pymongo.ASCENDING)],unique = True)
    

    
    
            
# def insert_map_results(map_name, algo):
#     if algo == "N":
#         author_name = "Jiaoyang"
#         algo_name = "CBSH2-RTC"
#     else:
#         author_name = "Bojie"
#         algo_name = "CBSH2-RTC + CHBP"
    
#     url = "mongodb+srv://MAPF:MONASH@mapfbenchmark.zsijqdc.mongodb.net/?retryWrites=true&w=majority"
#     client = pymongo.MongoClient(url)
#     db = client['MAPF_DB']
#     print("Connect successfully")
#     scen_type = ["even","random"]
#     col = db['mapf_instances']
#     lower_query = []
#     upper_query = []
#     solution_query = []
    
#     for i in range(1, 26):
#         for scen_t in scen_type: 
#             result_file = "../results/"+map_name+"/statistic/"+map_name+"-"+scen_t+"-"+str(i)+"-1800-"+algo+".csv"
#             print(result_file)
#             df1 = load_data(result_file)
#             res = col.find_one({"map_name": map_name ,"instance_id": i, "instance_type" :  scen_t})
#             instance_id = res['_id']
            
#             for index, row in df1.iterrows():
#                 if row['solution cost'] == -1:
#                     lower_bound = {
#                         "#agents":row['#agents'],
#                         "history":[
#                             {
#                             "algo_name": algo_name,
#                             "authors": author_name,
#                             "cost": row['min f value'],
#                             "runtime": row['runtime'],
#                             "date": str(datetime.datetime.now().date())
#                             }
#                         ]    
#                     }
#                     lower_query.append(UpdateOne({"map_instance_id": instance_id,"#agents": row['#agents']},
#                                                         {"$push": { "history": lower_bound['history'][0] }}
#                                                        ))
#                 else:
#                     history = {
#                             "algo_name": algo_name,
#                             "authors": author_name,
#                             "cost": row['solution cost'],
#                             "runtime": row['runtime'],
#                             "date": str(datetime.datetime.now().date())
#                             }
                    
#                     solution_query.append(UpdateOne({"map_instance_id": instance_id,"#agents": row['#agents']},
#                                     {"$push": { "history": history}}
#                                    ))
                      
#                     lower_query.append(UpdateOne({"map_instance_id": instance_id,"#agents": row['#agents']},
#                                     {"$push": { "history": history}}
#                                    ))
                    
#                     upper_query.append(UpdateOne({"map_instance_id": instance_id,"#agents": row['#agents']},
#                                     {"$push": { "history": history}}
#                                    ))
#     db['soultion_records'].bulk_write(solution_query)
#     print("Finish inserting solution")
#     db['lower_bound_records'].bulk_write(lower_query)
#     print("Finish inserting lower bound")
#     db['upper_bound_records'].bulk_write(upper_query)
#     print("Finish inserting upper bound")
    
    
# def clean_map(map_name):
#     url = "mongodb+srv://MAPF:MONASH@mapfbenchmark.zsijqdc.mongodb.net/?retryWrites=true&w=majority"
#     client = pymongo.MongoClient(url)
#     db = client['MAPF_DB']
#     print("Connect successfully")
#     db['soultion_records'].delete_many({ "history": { "$exists": True, "$size": 0 } })
#     db['lower_bound_records'].delete_many({ "history": { "$exists": True, "$size": 0 } })
#     db['upper_bound_records'].delete_many({ "history": { "$exists": True, "$size": 0 } })


# def upload_map_results(map_name):
#     add_all_agent_data(map_name)
#     insert_map_results(map_name, "N")
#     insert_map_results(map_name, "CHBP")
#     clean_map(map_name) 
```

```python
import datetime
# def data_field_exist(col,datafield)
def add_all_agent_data(map_name):    
    url = "mongodb+srv://MAPF:MONASH@mapfbenchmark.zsijqdc.mongodb.net/?retryWrites=true&w=majority"
    client = pymongo.MongoClient(url)
    db = client['MAPF']
    print("Connect successfully")
    col = db[map_name]
    scen_type = ["even","random"]  
    for i in range(1, 26):
        for scen_t in scen_type: 
            number_agent=col.find_one({"Instance_id": i, "Instance_type" :  scen_t})['Total_agent']
            all_agent_data=[]    
            for a in range(1,number_agent+1):
                all_agent_data.append(
                    {
                        "#Agents": a,
                        "History":[]
                    }
                )
            col.update_one({"Instance_id": i, "Instance_type" :  scen_t },
                        { 
                            "$set": {
                                "Solution": all_agent_data,
                                "Lower_bound": all_agent_data,
                                "Upper_bound": all_agent_data
                            }
                        }) 
            
def clean_empty(map_name):  
    url = "mongodb+srv://MAPF:MONASH@mapfbenchmark.zsijqdc.mongodb.net/?retryWrites=true&w=majority"
    client = pymongo.MongoClient(url)
    db = client['MAPF']
    print("Connect successfully")
    col = db[map_name]
    scen_type = ["even","random"]  
    for i in range(1, 26):
        for scen_t in scen_type: 
            col.update_many(
                { }, 
                {"$pull": {
                    "Solution": { "History": {"$size" : 0} },
                    "Lower_bound": { "History": {"$size" : 0} },
                    "Upper_bound": { "History": {"$size" : 0} }
                }
                })
    print("Finish cleaning")       
            
def insert_map_results(map_name, algo):
    if algo == "N":
        author_name = "Jiaoyang"
        algo_name = "CBSH2-RTC"
    else:
        author_name = "Bojie"
        algo_name = "CBSH2-RTC + CHBP"
    
    url = "mongodb+srv://MAPF:MONASH@mapfbenchmark.zsijqdc.mongodb.net/?retryWrites=true&w=majority"
    client = pymongo.MongoClient(url)
    db = client['MAPF']
    print("Connect successfully")
    col = db[map_name]
    scen_type = ["even","random"]
    
    for i in range(1, 26):
        for scen_t in scen_type: 
            result_file = "../results/"+map_name+"/statistic/"+map_name+"-"+scen_t+"-"+str(i)+"-1800-"+algo+".csv"
            print(result_file)
            df1 = load_data(result_file)
            for index, row in df1.iterrows():
                if row['solution cost'] == -1:
                    lower_bound = {
                        "#Agents":row['#agents'],
                        "History":[
                            {
                            "Algo_name": algo_name,
                            "Authors": author_name,
                            "Cost": row['min f value'],
                            "Runtime": row['runtime'],
                            "Date": str(datetime.datetime.now().date())
                            }
                        ]    
                    }
                    col.update_one({"Instance_id": i, "Instance_type" :  scen_t, "Lower_bound.#Agents": row['#agents'] },
                        { 
                            "$push": {
                                "Lower_bound.$.History": lower_bound['History'][0]
                            }
                        })
                else:
                    soultion = {
                        "#Agents":row['#agents'],
                        "History":[
                            {
                            "Algo_name": algo_name,
                            "Authors": author_name,
                            "Cost": row['solution cost'],
                            "Runtime": row['runtime'],
                            "Date": str(datetime.datetime.now().date())
                            }
                        ]    
                    }
                    lower_bound = {
                        "#Agents":row['#agents'],
                        "History":[
                            {
                            "Algo_name": algo_name,
                            "Authors": author_name,
                            "Cost": row['solution cost'],
                            "Runtime": row['runtime'],
                            "Date": str(datetime.datetime.now().date())
                            }
                        ]    
                    }
                    #inserting upper_bound data 
                    upper_bound = {
                        "#Agents":row['#agents'],
                        "History":[
                            {
                            "Algo_name": algo_name,
                            "Authors": author_name,
                            "Cost": row['solution cost'],
                            "Runtime": row['runtime'],
                            "Date": str(datetime.datetime.now().date())
                            }
                        ]    
                    }
                    col.update_one(
                        {"Instance_id": i, "Instance_type" :  scen_t },
                        {"$push": { "Solution.$[elem].History": soultion['History'][0],
                                    "Lower_bound.$[elem].History": lower_bound['History'][0],
                                    "Upper_bound.$[elem].History": upper_bound['History'][0]
                                  }},
                        upsert=False,
                        array_filters=[{"elem.#Agents": row['#agents']}]
                    )
    print("Finish inserting")    
#                     col.bulk_write([
#                          UpdateOne( {"Instance_id": i, "Instance_type" :  scen_t, "Solution.#Agents": row['#agents'] }, {"$push": { "Solution.$.History": soultion['History'][0]}}),
#                          UpdateOne( {"Instance_id": i, "Instance_type" :  scen_t, "Lower_bound.#Agents": row['#agents'] }, { 
#                             "$push": {
#                                 "Lower_bound.$.History": lower_bound['History'][0]
#                             }
#                          }),
#                          UpdateOne({"Instance_id": i, "Instance_type" :  scen_t, "Upper_bound.#Agents": row['#agents'] }, { "$push": { "Upper_bound.$.History": upper_bound['History'][0]}})
#                     ])

def upload_map_results(map_name):
    add_all_agent_data(map_name)
    insert_map_results(map_name,'CHBP')
    insert_map_results(map_name,'N')
    clean_empty(map_name)

```

```python
import datetime
# def data_field_exist(col,datafield)
def load_map_results(map_name, algo):
    if algo == "N":
        author_name = "Jiaoyang"
        algo_name = "CBSH2-RTC"
    else:
        author_name = "Bojie"
        algo_name = "CBSH2-RTC + CHBP"
    
    url = "mongodb+srv://MAPF:MONASH@mapfbenchmark.zsijqdc.mongodb.net/?retryWrites=true&w=majority"
    client = pymongo.MongoClient(url)
    db = client['MAPF']
    print("Connect successfully")
    print(db.list_collection_names())
    col = db[map_name]
    scen_type = ["even","random"]
    
    for i in range(1, 26):
        for scen_t in scen_type: 
            result_file = "../results/"+map_name+"/statistic/"+map_name+"-"+scen_t+"-"+str(i)+"-1800-"+algo+".csv"
            print(result_file)
            df1 = load_data(result_file)
            for index, row in df1.iterrows():
                if row['solution cost'] == -1:
                    lower_bound = {
                        "#Agents":row['#agents'],
                        "History":[
                            {
                            "Algo_name": algo_name,
                            "Authors": author_name,
                            "Cost": row['min f value'],
                            "Runtime": row['runtime'],
                            "Date": str(datetime.datetime.now().date())
                            }
                        ]    
                    }
                    
                    cur = col.find(
                        {
                            "Instance_id": i,
                            "Instance_type" :  scen_t,
                            "Lower_bound":{
                                "$elemMatch": {
                                     '#Agents': row['#agents']
                                }  
                            }
                        }
                    )
                    exist = list(cur)
                    if exist :
                        col.update_one({"Instance_id": i, "Instance_type" :  scen_t, "Lower_bound.#Agents": row['#agents'] },
                        { 
                            "$push": {
                                "Lower_bound.$.History": lower_bound['History'][0]
                            }
                        })
                    else:
                        col.update_one({"Instance_id": i, "Instance_type" :  scen_t},
                        { 
                            "$push": {
                                "Lower_bound": lower_bound
                            }
                        })
                else:
                    soultion = {
                        "#Agents":row['#agents'],
                        "History":[
                            {
                            "Algo_name": algo_name,
                            "Authors": author_name,
                            "Cost": row['solution cost'],
                            "Runtime": row['runtime'],
                            "Date": str(datetime.datetime.now().date())
                            }
                        ]    
                    }
                    cur = col.find(
                            {
                                "Instance_id": i,
                                "Instance_type" :  scen_t,
                                "Solution":{
                                    "$elemMatch": {
                                         '#Agents': row['#agents']
                                    }  
                                }
                            }
                        )
                    exist = list(cur)                    
                    if exist :
                        col.update_one({"Instance_id": i, "Instance_type" :  scen_t, "Solution.#Agents": row['#agents'] },
                        { 
                            "$push": {
                                "Solution.$.History": soultion['History'][0]
                            }
                        })
                    else:
                        col.update_one({"Instance_id": i, "Instance_type" :  scen_t},
                        { 
                            "$push": {
                                "Solution": soultion
                            }
                         })
                    #inserting lower_bound data 
                    lower_bound = {
                        "#Agents":row['#agents'],
                        "History":[
                            {
                            "Algo_name": algo_name,
                            "Authors": author_name,
                            "Cost": row['solution cost'],
                            "Runtime": row['runtime'],
                            "Date": str(datetime.datetime.now().date())
                            }
                        ]    
                    }
                    cur = col.find(
                            {
                                "Instance_id": i,
                                "Instance_type" :  scen_t,
                                "Lower_bound":{
                                    "$elemMatch": {
                                         '#Agents': row['#agents']
                                    }  
                                }
                            }
                        )
                    exist = list(cur)                    
                    if exist :
                        col.update_one({"Instance_id": i, "Instance_type" :  scen_t, "Lower_bound.#Agents": row['#agents'] },
                        { 
                            "$push": {
                                "Lower_bound.$.History": lower_bound['History'][0]
                            }
                        })
                    else:
                        col.update_one({"Instance_id": i, "Instance_type" :  scen_t},
                        { 
                            "$push": {
                                "Lower_bound": lower_bound
                            }
                         })
                        
                        
                    #inserting upper_bound data 
                    upper_bound = {
                        "#Agents":row['#agents'],
                        "History":[
                            {
                            "Algo_name": algo_name,
                            "Authors": author_name,
                            "Cost": row['solution cost'],
                            "Runtime": row['runtime'],
                            "Date": str(datetime.datetime.now().date())
                            }
                        ]    
                    }
                    cur = col.find(
                            {
                                "Instance_id": i,
                                "Instance_type" :  scen_t,
                                "Upper_bound":{
                                    "$elemMatch": {
                                         '#Agents': row['#agents']
                                    }  
                                }
                            }
                        )
                    exist = list(cur)                    
                    if exist :
                        col.update_one({"Instance_id": i, "Instance_type" :  scen_t, "Upper_bound.#Agents": row['#agents'] },
                        { 
                            "$push": {
                                "Upper_bound.$.History": upper_bound['History'][0]
                            }
                        })
                    else:
                        col.update_one({"Instance_id": i, "Instance_type" :  scen_t},
                        { 
                            "$push": {
                                "Upper_bound": upper_bound
                            }
                         })                        
                    
                    
    
```
